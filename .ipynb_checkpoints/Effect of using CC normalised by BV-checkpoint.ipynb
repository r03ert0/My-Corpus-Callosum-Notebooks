# Effect of using Corpus Callosum (CC) size normalised by Brain Volume (BV) size to control for differences in BV between cases and controls

*Roberto Toro, July 2014*

A frequent analysis strategy to control for differences in CC size among cases and controls is to divide CC by BV, that is, using relative instead of absolute CC volume. Here we study the effectivity of this strategy using simulations.

```python
%pylab inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas
```

## Constants
The following values are obtained from the ABIDE cohort.

```python
bv=1097252.27913964; # Brain volume
s=112310.67; # Standard deviation
a=10**(-0.409308); # intercept of allometric formula
b=0.6445876; # slope of allometric formula
v=0.19 # part of the variance in CC size explained by brain volume
iter=1000 # number of iterations
```

```python
def rnorm(n,u,s):
    return np.random.standard_normal(n)*s+u
def mean(x):
    return np.mean(x)
def sqrt(x):
    return np.math.sqrt(x)
def sd(x):
    return np.std(x)
```

## Simulation
We simulate groups of 50, 100, ... 350 subjects per group (variable n), with differences in average brain volume of 0.1, 0.2, ..., 1 standard deviations (variable d). Each time, we run 'iter' simulations, to compute the number of times that despite CC normalisation we are still able to observe a statistically significant difference in CC size.

```python
S=pandas.Series();
p=np.empty(iter);
for d in np.arange(1,0,-0.1):
    for n in np.arange(50,400,50):
        for i in range(0,iter):
            # brain volume size
            bv1=rnorm(n/2,bv-s*d/2,s);
            bv2=rnorm(n/2,bv+s*d/2,s);

            # part of the size of the corpus callosum related to brain volume (19% of the variance)
            cc1a=a*bv1**(b-1);
            cc2a=a*bv2**(b-1);

            # part of the size of the corpus callosum unrelated to brain volume (81% of the variance)
            cc1b=rnorm(n/2,mean(cc1a),sqrt(sd(cc1a)**2*(1/v-1)))
            cc2b=rnorm(n/2,mean(cc2a),sqrt(sd(cc2a)**2*(1/v-1)))

            # corpus callosum size
            cc1=cc1a+cc1b;
            cc2=cc2a+cc2b;

            y=np.concatenate((cc1,cc2));
            x=np.concatenate((bv1,bv2));
            g=np.concatenate((array([0]).repeat(n/2),array([1]).repeat(n/2)));

            df=pandas.DataFrame({'Y':y,'G':g});
            result=smf.ols(formula="Y ~ C(G)",data=df).fit();
            p[i]=result.f_pvalue;
        pw=(p<0.05).nonzero()[0].size;
        S=pandas.concat([S,pandas.Series([n,d,100*pw/iter])]);
D=pandas.DataFrame(S.reshape(70,3),columns=['N','D','Power']);

plt.figure();
for d in arange(0.1,1,0.1):
    D.loc[(D.D>d-0.05) & (D.D<d+0.05)].plot(x='N',y='Power')
plt.legend(np.arange(0.1,1.1,0.1))
```

The plot shows the power to detect a difference in corpus callosum induced by a difference in mean brain volume between groups of 1 standard deviation (top blue curve) to 0.1 standard deviations (bottom blue curve), for cohorts with 50 to 350 subjects per group.
